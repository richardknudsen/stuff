# Data structures

One largely distinguishes 3 different data types: 
- structured 
- semi-structured
- unstructured 

## Structured 

"Structured" data usually refers to tabular data. In the usual setup, data is organised in columns, where the values of each column are of one specific data type only. Regular data types include: 
- integers (e.g. 1, 2, 3) 
- floating-point (rational) numbers (e.g. 1.1, 2.14...)
- characters (e.g. 'some_key')
- boolean (True, False)

# SQL 

SQL (short for Structured Query Language) is THE standard high-level language to interact with databases, and was developed in the 60s. While it is mostly linked with relational databases, it's vocabulary is also used in most other types of databases, and provides a sort of standard vocabulary when talking about data manipulations. 

The exact syntax differs slightly between different SQL dialects and databases, but the major operations are mostly the same. 

In the following, we will quickly go over the standard query cases. For this purpose, we use a nice library called pandasql that emulates a simple SQLlite database. 

At first let us create two exemplary tables. 

```python
from pandasql import sqldf, load_meat, load_births


pysqldf = lambda q: sqldf(q, globals())
meat = load_meat()
births = load_births()
```

## Basic operations: SELECT, AS, FROM, LIMIT, WHERE, DISTINCT, AND, OR


Let's now checkout the first 5 rows of each table: 

```python
query = """
    SELECT * 
    FROM meat
    LIMIT 5
"""

pysqldf(query)
```

```python
query = """
    SELECT * 
    FROM births
    LIMIT 5
"""

pysqldf(query)
```

Let's now find our the date range of the births table: 

```python
query = """
    SELECT 
        min(date) as min_date, 
        max(date) as max_date
    FROM births
"""

pysqldf(query)
```

Let's assume we are now interested in finding the days where births were either below 250,000 or above 370,000: 

```python
query = """
    SELECT distinct(date)
    FROM births
    WHERE 
        births < 250000
        OR births > 370000
"""

pysqldf(query)
```

## Aggregations: GROUP BY, MIN, MAX, AVG, SUBQUERIES, 

Let's now find the 10 months with the highest number of births and compute the daily average: 

```python
query = """
    WITH births_with_month_and_year as (
        SELECT
            *, 
            strftime('%m', date) as month, 
            strftime('%Y', date) as year
        FROM births
    )
    SELECT 
        year, 
        month, 
        sum(births) as births_sum
    FROM births_with_month_and_year
    GROUP BY year, month 
    ORDER BY births DESC 
    LIMIT 10
"""

pysqldf(query)
```

## Combining tables: JOINs and UNIONs

```python
path = 'resources/sql-joins.png'
Image(filename=path)
```

Let's assume we want to get both births and meat statistics for those days where we have records for both: 

```python
query = """
    SELECT 
        b.date, 
        b.births, 
        m.beef
    FROM births b
    INNER JOIN meat m 
    ON b.date = m.date
"""

pysqldf(query)
```

Let's now split the meat table and put it back together: 

```python
query = """
    SELECT * 
    FROM meat
    WHERE date >= datetime('2000-01-01')
"""

meat1 = pysqldf(query)
meat1
```

```python
query = """
    SELECT * 
    FROM meat
    WHERE date < datetime('2000-01-01')
"""

meat2 = pysqldf(query)
meat2
```

```python
query = """
    SELECT * 
    from meat2
    UNION ALL
    SELECT * 
    FROM meat1
"""

meat_reconstructed = pysqldf(query)
meat_reconstructed
```
